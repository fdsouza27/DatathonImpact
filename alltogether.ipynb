```python
import requests

def get_repository_links():
    # URL of the repositories API endpoint
    url = "https://api.github.com/orgs/MIT-LCP/repos?type=all"

    # Send a GET request to retrieve the repositories
    response = requests.get(url)

    # Check the response status
    if response.status_code != 200:
        print("Error retrieving repositories:", response.text)
        return []  # Return an empty list if there is an error

    # Parse the JSON response
    repositories = response.json()

    # Extract the repository links
    repository_links = []
    for repo in repositories:
        repository_links.append(repo["html_url"])

    # Check if there are more pages
    while "next" in response.links:
        next_url = response.links["next"]["url"]
        response = requests.get(next_url)
        repositories = response.json()

        for repo in repositories:
            repository_links.append(repo["html_url"])

    return repository_links

# # Call the function and get the list of repository links
# repository_links = get_repository_links()

# # # Print the repository links
# for link in repository_links:
#     print(len(repository_links))

# Call the function and get the list of repository links
repository_links = get_repository_links()

# Print the repository links
for link in repository_links:
    print(link)
```

```python
import csv
import sys
import time
import pandas as pd
from scholarly import scholarly

def fetch_author_publications(target_author_name):
    retry_count = 0
    while retry_count < 3:  # Maximum number of retries
        try:
            search_query = scholarly.search_author(target_author_name)

            # Find the first author matching the target author name
            author_result = next(search_query)

            # Check if the author is found in the search results
            if author_result:
                author = scholarly.fill(author_result)

                # Print the author name
                print(f"Author Name: {target_author_name}")
                print()

                # Create a list to store publication details
                publication_details = []
                for publication in author['publications']:
                    publication_title = publication['bib'].get('title', '')
                    publication_year = publication['bib'].get('pub_year', '')
                    cited_by = publication.get('num_citations', '')
                    cited_by_url = publication.get('url_scholarbib', '')

                    # Get the abstract of the publication
                    abstract = scholarly.fill(publication).get('abstract', '')

                    # Get the authors of the publication
                    authors = publication.get('bib', {}).get('author', '')

                    publication_info = {
                        'Author Name': target_author_name,
                        'Publication Title': publication_title,
                        'Year': publication_year,
                        'Citations': cited_by,
                        'Cited By URL': cited_by_url,
                        'Abstract': abstract,
                        'Authors': authors
                    }

                    publication_details.append(publication_info)

                print("=======================================")

                # Convert the list to a DataFrame and return it
                return pd.DataFrame(publication_details)

        except StopIteration:
            print(f"Author '{target_author_name}' not found in search results.")
            return pd.DataFrame()

        except Exception as e:
            print(f"An error occurred: {str(e)}")
            retry_count += 1
            time.sleep(2)  # Delay between retries

    print("Max retries exceeded. Exiting.")
    return pd.DataFrame()

# Usage example
target_author_name = "SHUBHAM PACHORI"
publications_df = fetch_author_publications(target_author_name)

# Print the DataFrame as a proper table
display(publications_df)
```

```python
import pandas as pd
import plotly.express as px
from ipywidgets import interact, widgets, Layout, VBox, HBox
from IPython.display import display, HTML

# Assuming you have the DataFrame named publications_df from the previous scholarly code
main_dropdown_options = publications_df['Author Name'].unique().tolist()
second_dropdown_options = ['Before Datathon', 'After Datathon']
third_dropdown_options = ['All', 'First Author', 'Last Author']

def update_plot(main_selection, second_selection, third_selection):
    selected_data = publications_df[publications_df['Author Name'] == main_selection]

    if second_selection == 'Before Datathon':
        filtered_data = selected_data[selected_data['Year'] < 2014]
    elif second_selection == 'After Datathon':
        filtered_data = selected_data[selected_data['Year'] >= 2014]
    else:
        filtered_data = selected_data

    if third_selection == 'First Author':
        filtered_data = filtered_data[filtered_data['Authors'].str.startswith(main_selection)]
    elif third_selection == 'Last Author':
        filtered_data = filtered_data[filtered_data['Authors'].str.endswith(main_selection)]

    # Count the number of publications in each year
    publication_count = filtered_data.groupby('Year').size().reset_index(name='Count')

    # Replace any NaN values with zero
    publication_count['Count'].fillna(0, inplace=True)

    if filtered_data.empty:
        # Display a message if there are no publications for the selected author
        print(f"No publications found for {main_selection}.")
        return

    # Calculate the average number of publications per year as an integer (rounded off)
    avg_publications_per_year = int(publication_count['Count'].mean())

    bar_fig = px.bar(
        publication_count,
        x='Year',
        y='Count',
        labels={'Year': 'Year', 'Count': 'Publication Count'},
        text='Count',  # Display count on top of bars
        hover_data={'Year': False, 'Count': False},  # Hide hover for Year and Count
        hover_name='Year',  # Display year on hover
    )

    # Format the hover text to display the list of publication titles clearly
    bar_fig.update_traces(hovertemplate='Year: %{x}<br>Number of Publications: %{y}<br><br><b>Publication Titles:</b><br>%{customdata[0]}')

    # Convert the list of titles to a single string with each title on a separate line
    bar_fig.update_traces(customdata=[["<br>".join(titles)] for titles in filtered_data.groupby('Year')['Publication Title'].apply(list).values])

    # Set y-axis range to a fixed value to avoid scale changes
    y_range_max = publication_count['Count'].max() + 5  # Add some padding
    bar_fig.update_layout(yaxis=dict(range=[0, y_range_max]))

    bar_fig.update_layout(
        title=f"{main_selection} - Number of Publications vs Year (Average: {avg_publications_per_year})",
        xaxis=dict(title='Year'),
        yaxis=dict(title='Publication Count'),
    )

    bar_fig.show()

main_dropdown = widgets.Dropdown(
    options=main_dropdown_options,
    value=main_dropdown_options[0],
    description='Select author name:',
    layout=Layout(width='400px')
)

second_dropdown = widgets.Dropdown(
    options=second_dropdown_options,
    value=second_dropdown_options[0],
    description='Before/After Datathon:',
    layout=Layout(width='400px')
)

third_dropdown = widgets.Dropdown(
    options=third_dropdown_options,
    value=third_dropdown_options[0],
    description='Select author position:',
    layout=Layout(width='400px')
)

# Create a label for the main title using HTML styling
main_title = widgets.HTML("<h1 style='font-weight:bold;'>Datathon Attendees Dashboard</h1>")

# Create labels for each dropdown using HTML styling
main_dropdown_title = widgets.HTML("<h3 style='font-weight:bold;'>Author Name:</h3>")
second_dropdown_title = widgets.HTML("<h3 style='font-weight:bold;'>Before/After Datathon:</h3>")
third_dropdown_title = widgets.HTML("<h3 style='font-weight:bold;'>Author Position:</h3>")

# Use HBox to display dropdowns and their titles side by side
main_dropdown_box = VBox([main_dropdown_title, main_dropdown])
second_dropdown_box = VBox([second_dropdown_title, second_dropdown])
third_dropdown_box = VBox([third_dropdown_title, third_dropdown])

# Use interactive_output to connect the update_plot function with the dropdowns
output = widgets.interactive_output(update_plot, {'main_selection': main_dropdown, 'second_selection': second_dropdown, 'third_selection': third_dropdown})

# Display the widgets and output together
display(main_title, HBox([main_dropdown_box, second_dropdown_box, third_dropdown_box]), output)
```
